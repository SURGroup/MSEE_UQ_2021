    - $f(\theta)=\theta_{0} x + \theta_{1} x^{2} + \theta_{2} x^{3}$

   where $\theta$ are the parameters of each model. You are given the file 'data_ex1a.txt' which contains 50 synthetic measurements generated from one of these models and some added noise.
